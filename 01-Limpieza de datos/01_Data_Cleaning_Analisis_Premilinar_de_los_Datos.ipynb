## Missing Values

Hay valores faltantes en la columna Age y body




```python
pd.isnull(data['age'])
```

```python
pd.isnull(data['body'])
```

```python
pd.isnull(data['body']).values.ravel().sum() # Suma la cantidad de valores true ( Valores nulos )
```

```python
pd.notnull(data['body']).values.ravel().sum() # Suma la cantidad de valores false ( Valores no nulos )
```

Razones por las que los valores faltan en un dataset:
* Extracción de los datos
* Recolección de los datos

### Borrado de valores que faltan

```python
data.dropna( axis = 0, how = "all" ) # axis = 1 borra la columna, all: borra la fila si todas las columnas tienen na
```

```python
data2 = data
data2.dropna( axis = 0, how = "any") # Elimina la fila con solo un na 
```

### Cómputo de los valores faltantes

```python
data3 = data
data3.fillna(0) # cambia los na por 0
```

```python
data4 = data
data4.fillna('Desconocido')
```

```python
data5 = data
data5['body'] = data5['body'].fillna(0)
data5['home.dest'] = data5['home.dest'].fillna('Desconocido')
data5
```

```python
pd.isnull(data5['age']).values.ravel().sum()
```

```python
data5['age'] = data5['age'].fillna(data['age'].mean()) # Reemplazamos los na con el promedio
data5
```

```python
data5.age.fillna(method='ffill') # rellena los na con el valor no na mas cercano que este hacia adelante
data5.age.fillna(method='backfill') # rellena los na con el valor no na mas cercano que este hacia atras
```

Crearemos dos columnas una llamada male y otr female 

```python
dummy_sex = pd.get_dummies(data['sex'], prefix = 'sex')
dummy_sex
```

```python
column_name = data.columns.values.tolist()
column_name
```

```python
data.drop(['sex'], axis = 1) # Se elimina la columna sex
```

```python
data = pd.concat([data, dummy_sex], axis = 1)
```

```python
def createDummies( df, var_name ):
  dummy = pd.get_dummies(df[var_name], prefix = var_name)
  df = df.drop(var_name, axis = 1)
  df = pd.concat([df, dummy ], axis = 1)
  return df
```

```python
createDummies( data, 'sex' )
```
